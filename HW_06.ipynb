````python
# TODO: Adjust the path below as necessary. This assumes the file is in the same directory as this script.

SOWPODS_PATH = "sowpods.txt"

def read_sowpods(fpath: str) -> list[str]:
    """Read the SOWPODS dictionary file and return a set of words."""
    with open(fpath, "r") as f:
        return [line.strip("\n") for line in f.readlines()]


print(f"Read {len(read_sowpods(SOWPODS_PATH))} words from SOWPODS file.")

scores = {
    "a": 1, "c": 3, "b": 3, "e": 1, "d": 2, "g": 2,
    "f": 4, "i": 1, "h": 4, "k": 5, "j": 8, "m": 3,
    "l": 1, "o": 1, "n": 1, "q": 10, "p": 3, "s": 1,
    "r": 1, "u": 1, "t": 1, "w": 4, "v": 4, "y": 4,
    "x": 8, "z": 10
}

WORDS = read_sowpods(SOWPODS_PATH)
````

````python
# Q6-2 Grading Tag:


def score_word(word: str) -> int:
    """
    Calculates the Scrabble score of a word.
    Normal letters use the score dictionary and wildcards are worth zero points.
    """
    total = 0

    # Check each character and add its Scrabble value
    for letter in word.lower():
        if letter in scores:
            total += scores[letter]
        # A wildcard contributes no points
        elif letter == "*":
            total += 0

    return total

def validate_rack(rack: str) -> bool:
    """
    Checks if a Scrabble rack is valid.
    A valid rack has 2-7 characters and only contains letters or *.
    """

    # Rack size must follow normal Scrabble rules
    if len(rack) < 2 or len(rack) > 7:
        return False

    wildcard_count = 0

    # Check every character in the rack
    for char in rack:
        if char == "*":
            wildcard_count += 1

        # Allow uppercase and lowercase letters only
        elif not char.isalpha() or not char.isascii():
            return False

    # A rack can have at most two blank tiles
    if wildcard_count > 2:
        return False

    return True

def run_scrabble(rack: str, words: list[str]) -> dict[str, int]:
    """
    Finds all words that can be created from the rack.
    Returns a dictionary containing valid words and their scores.
    """

    result = {}

    rack = rack.upper()

    # Count the letters available in the rack
    rack_letters = {}

    wildcards = 0

    for letter in rack:
        if letter == "*":
            wildcards += 1
        else:
            rack_letters[letter] = rack_letters.get(letter, 0) + 1

    for word in words:

        word = word.upper()

        # Ignore words longer than the rack
        if len(word) > len(rack):
            continue

        needed_letters = {}
        possible = True

        # Count the letters required for this word
        for letter in word:
            needed_letters[letter] = needed_letters.get(letter, 0) + 1

        missing = 0

        # Check how many letters must be supplied by wildcards
        for letter, amount in needed_letters.items():
            available = rack_letters.get(letter, 0)

            if amount > available:
                missing += amount - available

        # If missing letters are more than wildcards, cannot make word
        if missing > wildcards:
            continue

        # This logic keeps the highest scoring version.
        # It is robust because it automatically handles cases where
        # a wildcard could replace a normal tile.
        normal_score = 0

        remaining_letters = rack_letters.copy()

        for letter in word:
            if remaining_letters.get(letter, 0) > 0:
                normal_score += scores[letter.lower()]
                remaining_letters[letter] -= 1

        # Remaining letters in the word are covered by wildcards
        result[word] = max(result.get(word, 0), normal_score)

    return result

if __name__ == "__main__":

    words = read_sowpods(SOWPODS_PATH)

    rack = "EXAMPLE"

    # First check if the rack is valid
    if validate_rack(rack):
        result = run_scrabble(rack, words)
        print(result)
    else:
        print("Invalid rack")
````
