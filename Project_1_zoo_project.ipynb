````python
# QP1-1 Grading Tag:

# TODO: ALL OF YOUR CODE TO GRADE NEEDS TO EXIST IN THIS CELL
import random
from abc import ABC, abstractmethod

# LocalView Class

class LocalView:
    """
    Represents the directions an animal is allowed to move.
    The Zoo creates this object and passes it to the Animal.
    """

    def __init__(self, can_move_north, can_move_east,
                 can_move_south, can_move_west):

        # to stay in place
        self.move_directions = ["stay"]

        if can_move_north:
            self.move_directions.append("north")
        
        if can_move_east:
            self.move_directions.append("east")
        
        if can_move_south:
            self.move_directions.append("south")

        if can_move_west:
            self.move_directions.append("west")

# Animal Abstract Base Class

class Animal(ABC):
    """
    Parent class for all animals.

    Every animal:
    - has an icon
    - can be printed
    - must implement its own move() method
    """
    # Included for posterity.

    def __init__(self, icon):
        self.icon = icon

    @property
    def icon(self):
        return self._icon

    @icon.setter
    def icon(self, value):
        # The assignment requires the icon to be exactly one character
        if len(value) != 1:
            raise Exception("icon must be a single character.")
        self._icon = value

    def __str__(self):
        return self.icon

    @abstractmethod
    def move(self, local_view):
        """
        Decide which direction to move.
        Must be implemented by every subclass.
        """
        pass

# Lion Class

class Lion(Animal):
    """
    Lion moves randomly from the available directions.
    """

    def __init__(self):
        super().__init__("🦁")

    def move(self, local_view):
        return random.choice(local_view.move_directions)


# Peacock Class

class Peacock(Animal):
    """
    Peacock prefers moving instead of staying.
    If no movement is possible, it stays.
    """

    def __init__(self):
        super().__init__("🦚")

    def move(self, local_view):

        # If there is at least one movement option,
        # choose one of them instead of staying.
        if len(local_view.move_directions) > 1:
            return random.choice(local_view.move_directions[1:])

        return "stay"

# Zoo Class

class Zoo:
    """
    Represents a zoo where animals move around on a grid.
    """

    def __init__(self, height, width):
        """
        Create a zoo with the given height and width.
        """
        self.height = height
        self.width = width

        # Dictionary to store animal positions
        # Key   -> Animal object
        # Value -> (row, column)
        self.animals = {}

    def add_animal(self, animal, row, column):
        """
        Add an animal to the zoo at the specified position.
        """

        # Check if row and column are within the zoo boundaries
        if row < 0 or row >= self.height:
            raise ValueError("Row is outside the zoo.")

        if column < 0 or column >= self.width:
            raise ValueError("Column is outside the zoo.")

        # Check if another animal already occupies this position
        if self.animal_at_position(row, column) is not None:
            raise ValueError("Position is already occupied.")

        # Store the animal and its position
        self.animals[animal] = (row, column)

    def position_of_animal(self, animal):
        """
        Return the (row, column) position of the given animal.
        Returns None if the animal is not in the zoo.
        """
        return self.animals.get(animal)

    def animal_at_position(self, row, column):
        """
        Return the animal at the given position.
        Returns None if the position is empty.
        """

        for animal, position in self.animals.items():
            if position == (row, column):
                return animal

        return None
    
    def _new_position(self, row, column, direction):
        """
        Return the new position after moving in the given direction.
        """

        if direction == "north":
            return (row - 1, column)

        if direction == "south":
            return (row + 1, column)

        if direction == "east":
            return (row, column + 1)

        if direction == "west":
            return (row, column - 1)

        return (row, column)

    def time_step(self):
        """
        Give every animal one chance to move.
        Animals move one at a time.
        """

        # Make a copy so we can safely update positions
        animals = list(self.animals.keys())

        for animal in animals:

            row, column = self.animals[animal]

            # Check which directions are available

            can_move_north = (
                row > 0 and
                self.animal_at_position(row - 1, column) is None
            )

            can_move_south = (
                row < self.height - 1 and
                self.animal_at_position(row + 1, column) is None
            )

            can_move_east = (
                column < self.width - 1 and
                self.animal_at_position(row, column + 1) is None
            )

            can_move_west = (
                column > 0 and
                self.animal_at_position(row, column - 1) is None
            )

            local_view = LocalView(
                can_move_north,
                can_move_east,
                can_move_south,
                can_move_west
            )

            direction = animal.move(local_view)

            new_row, new_column = self._new_position(
                row,
                column,
                direction
            )

            self.animals[animal] = (new_row, new_column)
        
    def __str__(self):
        """
        Return a string representation of the zoo.
        """

        rows = []

        border = "+" + "-" * self.width + "+"
        rows.append(border)

        for row in range(self.height):

            line = "|"

            for column in range(self.width):

                animal = self.animal_at_position(row, column)

                if animal is None:
                    line += "."

                else:
                    line += str(animal)

            line += "|"
            rows.append(line)

        rows.append(border)

        return "\n".join(rows)
````
